```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer as NB
```

```python
data = []
with open('assignment 3 data.txt', 'r') as readfile:
    raw = readfile.readlines()
    for text in raw:
        data.append(text.strip())
```

```python
data
```

```python
sentiment_lst = []
patt_pos = 0
patt_neg = 0
nb_count_pos = 0
nb_count_neg = 0

for text in data:
    print(text)
    polar = TextBlob(text).sentiment.polarity
    if polar >= 0.33:
        sent = 'Positive'
    elif polar <= -0.33:
        sent = "Negative"
    else:
        sent = "Neutral"
    nbayes = TextBlob(text, analyzer = NB())
    sentiment_lst.append((sent, nbayes.sentiment.classification))
    if polar >= 0:
        patt_pos += 1
    else:
        patt_neg += 1
    if nbayes.sentiment.classification == 'pos':
        nb_count_pos += 1
    else:
        nb_count_neg += 1
    print(f"Pattern analyzer\nPolarity: {polar}\t\tSentiment: {sent}")
    print(f"Naive Bayes analyzer\nProbability: {nbayes.sentiment.p_pos}\t\tSentiment: {nbayes.sentiment.classification}\n")
```

Using the pattern analyzer and labeling text as positive if the polarity is >= 0 we can compare the pattern analyzer and the naive Bayes analyzer results below. Naive Bayes did not have a neutral rating so neutral is not counted.

```python
import pandas as pd
df = pd.DataFrame({"Pattern analyzer": [patt_pos,patt_neg], "Naive Bayes":[nb_count_pos,nb_count_neg]})
df.index = ["Positive", "Negative"]
df
```

### Summary

The results are interesting. The pattern analyzer labeled 3/4 of the data as positive while naive Bayes found just over 50% positive. Without a labeled training and testing set it would be difficult to determine which analyzer performed better. Below I'll examine where the two classifiers gave opposite labels, such as the pattern analyzer being negative and the naive bayes analyzer being positive.

```python
fix_sentiment_lst = [('pos', x[1]) if x[0] == 'Positive' else x for x in sentiment_lst]
fix_sentiment_lst = [('neg', x[1]) if x[0] == 'Negative' else x for x in fix_sentiment_lst]

for i, tup in enumerate(fix_sentiment_lst):
    if tup[0] != 'Neutral':
        if tup[0] != tup[1]:
            print(data[i])
            print(f"Pattern Analyzer: {tup[0]}\nNaive Bayes: {tup[1]}\n")
```

From the 4 opposite classifications it appears that the pattern analyzer is doing a better job than the naive Bayes analyzer. I would personally label all 4 of these as positive but the naive Bayes labeled them negative.

Looking back up above in the original analysis print out we can see that the last text analyzed "The portion was huge!" was labeled negative by the pattern analyzer at 0.48, which in my opinion should be neutral because it is so close to the middle of the scoring system. These are all small nuances that should be carefully understood before applying a machine learning tool to text analysis.
